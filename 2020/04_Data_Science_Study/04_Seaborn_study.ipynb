### Parte 1 - Distribution Plots

```python
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Essa função abaixo é a soma de todas as gaussiana
#perceba que no fim temos um pequeno pico,
#pois o número 50 aparece 2x
```

```python
sns.distplot(tips['total_bill'])
```

```python
tips['total_bill'].mean()
```

```python
sns.distplot(tips['total_bill'],kde = False)
```

```python
sns.distplot(tips['total_bill'],bins = 40)
#quantidade de histogramas
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind ='hex' )
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind ='reg' )
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind ='kde' )
```

```python
data1 = {
    'idade':[1,2.5,6,11,5.5,1.5,2,20,30],
    'gasto':[40,30,20,10,1,1,10,10,15]
}

df = pd.DataFrame(data = data1)
sns.distplot(df['idade'])
```

```python
sns.jointplot(x='gasto',y='idade',data=data1, kind ='kde' )
```
