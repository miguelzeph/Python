```python
tips.head(4)
```

```python
tips['sex']

h = 0
m = 0

for i in range(0,len(tips['sex'])):
    if tips['sex'][i] == 'Male':
        h = h + 1
    else:
        m = m + 1
print(
    'mulher: %3.0f\n'%m,
    'homem: %3.0f\n'%h,
    'total: %3.0f'%(h+m)
)
```

### Parte 3 - Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)

#Plotar Mapas na Diagonais
g.map_diag(sns.distplot)

#Plotar Mapas superior
g.map_upper(plt.scatter)

#Plotar Mapas Inferiores
g.map_lower(sns.kdeplot)
```

### Parte 4 - Regression Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data = tips )
```

```python
#hue = comparar
sns.lmplot(
    x='total_bill',
    y='tip',
    data = tips,
    hue ='sex',#Compara
    markers = ['*','v'],#Muda o plot
    scatter_kws = {'s':150}#Aumenta o tamanho
)
```

```python
sns.lmplot(
    x='total_bill',
    y='tip',
    data = tips,
    col = 'day',
    #row='time',
    hue = 'sex',
    aspect = 0.6,
    #size =8 #mudou o nome pra height
    height = 8,   
)
#plt.savefig('./teste.png')
```
