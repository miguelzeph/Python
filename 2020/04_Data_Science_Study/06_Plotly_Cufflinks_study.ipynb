### Box plot - Explicação 

Uma análise das estatísticas descritivas da amostra é fundamental para resumirmos algumas informações sobre a população. Estas informações são utilizadas para tomada de decisão e formação de modelos estatísticos paramétricos. Definiremos como:

Mínimo: menor elemento da amostra;

Máximo: maior elemento da amostra;

Quartis (Q1, Q2 e Q3): São valores dados a partir do conjunto de observações ordenado em ordem crescente, que dividem a distribuição em quatro partes iguais. O primeiro quartil, Q1, é o número que deixa 25% das observações abaixo e 75% acima, enquanto que o terceiro quartil, Q3, deixa 75% das observações abaixo e 25% acima. Já Q2 é a mediana, deixa 50% das observações abaixo e 50% das observações acima.

fonte: http://www.portalaction.com.br/estatistica-basica/23-quartis#quartis

```python
df.iplot(kind = 'box')
```

```python
# Confirmar os dados A do boxplot

# Mas primeiro...Passar dados pra ordem crescente (não pode esquecer)

df_crescente = df['A'].sort_values() #passar para ordem crescente!!!

#porém os índices não variam... continuam conforme a forma antiga, então, devemos resetá-los
df_crescente = df_crescente.reset_index()

df_crescente.head(4)
```

```python
df['A'].sort_values().reset_index()['A'].head(4)
```

```python
#Calcular os valores

#índice do quartil 1
q1 = int((len(df['A'])+1)/4.0)

# ''''''
q3 = int(3*(len(df['A'])+1)/4.0)

print(
    'Max:%.5f\n'%df['A'].max(),
    'Terceiro Quartil(Q3): %.5f\n'%(df_crescente['A'][q3-1]),# não deu
    'Median ou segundo quartil (Q2):%.5f\n'%df['A'].median(),
    'Primeiro Quartil(Q1): %.5f\n'%(df_crescente['A'][q1-1]),# não deu
    'Min:%.5f\n'%df['A'].min(),
)
```
