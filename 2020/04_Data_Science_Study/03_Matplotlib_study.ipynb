# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'-r')
plt.xlabel('Frequencia')
plt.ylabel('Intensidade')

plt.subplot(1,2,2)
plt.plot(y,x,'-b')
plt.xlabel('Frequencia')
plt.ylabel('Intensidade')

plt.show()
```

```python
# Podemos n√≥s mesmo desenhar os eixos

fig = plt.figure() #Tela em branco

#(POS X, POS Y, LARGURA,ALTURA)
axes = fig.add_axes([0.1,1,0.5,1])


axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.45,0.55,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("LARGER")
axes1.set_xlabel('Frequencia')
axes1.set_ylabel('Intensidade')

axes2.plot(y,x)
axes2.set_title("SMALLER")
axes2.set_xlabel('Frequencia')
axes2.set_ylabel('Intensidade')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```
