```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,10,11,endpoint= True)
y = np.cos(-x**2/9.0) + x/4
f = interp1d(x,y) # Kind = Linear é o Padrão (standard)
f2 = interp1d(x,y,kind= 'cubic')

xnew = np.linspace(0,10,41,endpoint = True)
```

<font size = 4>$f(x) = cos(\frac{-x^{2}}{9})+ \frac{x}{4}$</font>

```python
plt.figure(figsize = (7,5))
plt.plot(x,y,'o',xnew,f(xnew),'-',xnew,f2(xnew),'--')
plt.legend(
    ['Real data','Linear Interpolation','Cubic Interpolation'],
    loc = 'best',
)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Interpolação',fontsize = 15);
```

```python
x = np.linspace(0,10,11,endpoint = True)
y = np.cos(-x**2/9.0) + x/4

f1 = interp1d(x,y,kind='nearest')
f2 = interp1d(x,y,kind='previous')
f3 = interp1d(x,y,kind='next')
```

$f(x,y) = x(1-x). cos(4\pi x) .sen^{2}(4 \pi y^{2})$

```python
def func(x,y):
    return x*(1-x)*np.cos(4*np.pi*x) * np.sin(4*np.pi*y**2)**2
```

```python
grid_x, grid_y = np.mgrid[0:1:50j,0:1:50j]
```

```python
points = np.random.rand(500,2)
values = func(points[:,0],points[:,1])
```

```python
from scipy.interpolate import griddata
```

```python
#Interpolação
grid_z0 = griddata(
    points,
    values,
    (grid_x,grid_y),
    method = 'nearest'
    
)
grid_z1 = griddata(
    points,
    values,
    (grid_x,grid_y),
    method = 'linear'
    
)
grid_z2 = griddata(
    points,
    values,
    (grid_x,grid_y),
    method = 'cubic'
    
)
```

```python
plt.subplots_adjust(
    left=0.125,
    bottom=0.1,
    right=0.9,
    top=0.9,
    wspace=0.2,
    hspace=0.5
)

plt.subplot(221)
plt.imshow(
    func(grid_x,grid_y).T,
    extent = (0,1,0,1),
    origin = 'lower'
)
#plt.plot(points[:,0],points[:,1],'.k',ms=1)
plt.title('Original')


plt.subplot(222)
plt.imshow(
    grid_z0.T,
    extent = (0,1,0,1),
    origin = 'lower'
)
plt.title('Nearest')


plt.subplot(223)
plt.imshow(
    grid_z1.T,
    extent = (0,1,0,1),
    origin = 'lower'
)
plt.title('Linear')


plt.subplot(224)
plt.imshow(
    grid_z2.T,
    extent = (0,1,0,1),
    origin = 'lower'
)
plt.title('Cubic')

```

```python
from matplotlib import cm

#Using radial basis functions for smoothing/interpolation
from scipy.interpolate import Rbf
```

```python
x = np.random.rand(100)*4-2
y = np.random.rand(100)*4-2
z = x*np.exp(-x**2-y**2)
ti = np.linspace(-2,2,100)
XI,YI = np.meshgrid(ti,ti)
```

```python
rbf = Rbf(x,y,z,epislon =2,function ='multiquadric' )#multiquadric DEFAULT
ZI = rbf(XI,YI)
```

```python

plt.pcolor(XI, YI, ZI, cmap=cm.jet)
plt.scatter(x, y, 100, z, cmap=cm.jet)
plt.title('RBF interpolation - multiquadrics')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.colorbar()
```
