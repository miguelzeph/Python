Linha (y) e coluna (x) ... ou seja a linha e coluna indica a posio dos pixels...  temos 130 linhas e em cada linha temos 130 colunas e em cada coluna temos 3 camadas. Essas 3 camadas tem valores de 0(preto) at 255 (Claro)... esses valores se misturam para formar uma COR.

Portanto, as linhas e colunas so a posio espacial dos pontos no grfico, e a cor  defenida pela mistura dos valores das 3 camadas RGB.. (ex: [0,0,255] = AZUL).
